```python
def plot_2d_svc_problem(X, y, svc=None):
    '''
    Plots a two-dimensional labeled dataset (X,y) and, if SVC object is given, 
    the decision surfaces (with margin as well).
    '''
    assert X.shape[1] == 2, "Dataset is not two-dimensional"
    if svc!=None : 
        # Create a mesh to plot in
        r = 0.03  # mesh resolution
        x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
        y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
        xx, yy = np.meshgrid(np.arange(x_min, x_max, r),
                             np.arange(y_min, y_max, r))
        XX=np.c_[xx.ravel(), yy.ravel()]
        Z = np.array([svc_predict(svc, x) for x in XX])
        # Put the result into a color plot
        Z = Z.reshape(xx.shape)
        plt.contourf(xx, yy, Z, cmap=plt.cm.Pastel1)

    # Plot the dataset
    plt.scatter(X[:,0],X[:,1], c=y, cmap=plt.cm.Paired, marker='o', s=50)

def svc_predict(svc, x) : 
    h = svc.decision_function([x])
    if np.isclose(h, 0, atol=0.03):
        return 5
    elif (h >= -1 and h < -0.03) or (h > 0.03 and h <= 1):
        return 0.5
    else: 
        return max(-1, min(1, h))
    
def plot_error_surface(err, c_range=(0,5), g_range=(0,5)):
    c1, c2 = c_range[0], c_range[1]
    g1, g2 = g_range[0], g_range[1]
    plt.xticks(range(0,g2-g1+1,5),range(g1,g2+1,5)); plt.xlabel("gamma")
    plt.yticks(range(0,c2-c1+1,5),range(c1,c2+1,5)); plt.ylabel("C")
    p = plt.contour(err);
    plt.imshow(1-err, interpolation='bilinear', origin='lower',cmap=plt.cm.gray)
    plt.clabel(p, inline=1, fontsize=10)
    
```

Također, izračunajte prosječni gubitak SVM-a na skupu `seven`. Uvjerite se da je rezultat identičan onome koji biste dobili primjenom ugrađene funkcije [`metrics.hinge_loss`](http://scikit-learn.org/stable/modules/generated/sklearn.metrics.hinge_loss.html).

```python
from sklearn.metrics import hinge_loss

def hinge(model,x,y):
    return max(0,(1-y*model.decision_function(x))[0])

print(f'Gubitak modela za x2=(3,2) koji je označen pozitivno : {hinge(clf,[[3,2]],1)}')
print(f'Gubitak modela za x1=(3.5,2) koji je označen pozitivno : {hinge(clf,[[3.5,2]],1)}')
print(f'Gubitak modela za x=(4,2) koji je označen negativno : {hinge(clf,[[4,2]],-1)}')

def avg_hinge(model,X,y):
    sum=0
    N=len(X)
    for i in range(N):
        sum+=hinge(model,[X[i]],y[i])
    return sum/N

print(f'Prosječni gubitak SVM-a : {avg_hinge(clf,seven_X,seven_y)}')

y_pred= clf.decision_function(seven_X)

print(f'Prosječni gubitak SVM-a uporabom funkcije hinge_loss : {hinge_loss(seven_y,y_pred)}')
```

### 3. Optimizacija hiperparametara SVM-a

Pored hiperparametra $C$, model SVM s jezgrenom funkcijom RBF ima i dodatni hiperparametar $\gamma=\frac{1}{2\sigma^2}$ (preciznost). Taj parametar također određuje složenost modela: velika vrijednost za $\gamma$ znači da će RBF biti uska, primjeri će biti preslikani u prostor u kojem su (prema skalarnome produktu) međusobno vrlo različiti, što će rezultirati složenijim modelima. Obrnuto, mala vrijednost za $\gamma$ znači da će RBF biti široka, primjeri će biti međusobno sličniji, što će rezultirati jednostavnijim modelima. To ujedno znači da, ako odabremo veći $\gamma$, trebamo jače regularizirati model, tj. trebamo odabrati manji $C$, kako bismo spriječili prenaučenost. Zbog toga je potrebno zajednički optimirati hiperparametre $C$ i $\gamma$, što se tipično radi iscrpnim pretraživanjem po rešetci (engl. *grid search*). Ovakav pristup primjenjuje se kod svih modela koji sadrže više od jednog hiperparametra.


#### (a)

Definirajte funkciju 

> `grid_search(X_train, X_validate, y_train, y_validate, c_range=(c1,c2), g_range=(g1,g2), error_surface=False)` 

koja optimizira parametre $C$ i $\gamma$ pretraživanjem po rešetci. Funkcija treba pretražiti hiperparametre $C\in\{2^{c_1},2^{c_1+1},\dots,2^{c_2}\}$ i $\gamma\in\{2^{g_1},2^{g_1+1},\dots,2^{g_2}\}$. Funkcija treba vratiti optimalne hiperparametre $(C^*,\gamma^*)$, tj. one za koje na skupu za provjeru model ostvaruju najmanju pogrešku. Dodatno, ako je `surface=True`, funkcija treba vratiti matrice (tipa `ndarray`) pogreške modela (očekivanje gubitka 0-1) na skupu za učenje i skupu za provjeru. Svaka je matrica dimenzija $(c_2-c_1+1)\times(g_2-g_1+1)$ (retci odgovaraju različitim vrijednostima za $C$, a stupci različitim vrijednostima za $\gamma$).

```python
from sklearn.metrics import accuracy_score, zero_one_loss

def grid_search(X_train, X_validate, y_train, y_validate, c_range=(0,5), g_range=(0,5), error_surface=False):
    Matrix_test=np.ndarray(shape=(len(c_range)+1,len(g_range)+1))
    Matrix_train=np.ndarray(shape=(len(c_range)+1,len(g_range)+1))
    best_c=0
    best_g=0
    min_loss=-1
    for c in c_range:
        i=c_range.index(c)
        for g in g_range:
            j=g_range.index(g)

            clf=SVC(C=2**c,gamma=2**g)
            clf.fit(X_train,y_train)
            loss_train=zero_one_loss(y_train,clf.predict(X_train))
            loss_test=zero_one_loss(y_validate,clf.predict(X_validate))

            Matrix_train[i][j]=loss_train
            Matrix_test[i][j]=loss_test

            if loss_test < min_loss or min_loss==-1:
                min_loss=loss_test
                best_g=g
                best_c=c

    if error_surface:
        return best_c,best_g,Matrix_train,Matrix_test

    return best_c,best_g
```

```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split

X2, y2 = make_classification(n_samples=200,n_features=2,n_classes=2,n_redundant=0,n_clusters_per_class=2)
X100, y100 = make_classification(n_samples=200,n_features=100,n_classes=2,n_redundant=0,n_clusters_per_class=2)

X2_train,X2_test, y2_train,y2_test =train_test_split(X2, y2, test_size=0.5)
X100_train,X100_test, y100_train,y100_test =train_test_split(X100, y100, test_size=0.5)

c_range=range(-5,15+1)
g_range=range(-15,3+1)

best_c2,best_g2,Matrix_train2,Matrix_test2=grid_search(X2_train,X2_test,y2_train,y2_test,c_range,g_range,True)
best_c100,best_g100,Matrix_train100,Matrix_test100=grid_search(X100_train,X100_test,y100_train,y100_test,c_range,g_range,True)

print(f'Optimalni c za set s 2 dimenzije : {best_c2}')
print(f'Optimalni g za set s 2 dimenzije : {best_g2}')

print(f'Optimalni c za set s 100 dimenzija : {best_c100}')
print(f'Optimalni g za set s 2 dimenzije : {best_g100}')

#print(Matrix_test100)

figure(1)
plot_error_surface(Matrix_train2,c_range,g_range)
figure(2)
plot_error_surface(Matrix_test2,c_range,g_range)
figure(3)
plot_error_surface(Matrix_train100,c_range,g_range)
figure(4)
plot_error_surface(Matrix_test100,c_range,g_range)
```

```python
from statistics import mean

standard_accuracies_train=0
minmax_accuracies_train=0
regular_accuracies_train=0

standard_accuracies_test=0
minmax_accuracies_test=0
regular_accuracies_test=0

iter=50

for i in range(iter):
    X, y = make_classification(n_samples=500,n_features=2,n_classes=2,n_redundant=0,n_clusters_per_class=1)
    X[:,1] = X[:,1]*100+1000
    X[0,1] = 3000
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5)

    #--------------- Neskalirane značajke---------------------#
    clf=SVC(kernel='rbf').fit(X_train,y_train)
    accuracy_train=accuracy_score(y_train,clf.predict(X_train))
    accuracy_test=accuracy_score(y_test,clf.predict(X_test))

    regular_accuracies_train+=accuracy_train
    regular_accuracies_test+=accuracy_test

    #--------------- Standardizirane značajke-----------------#
    standard_scaler = StandardScaler()
    standard_scalered_x_train=standard_scaler.fit_transform(X_train)
    standard_scalered_x_test=standard_scaler.transform(X_test)

    standard_clf=SVC().fit(standard_scalered_x_train,y_train)
    standard_scalerd_accuracy_train=accuracy_score(y_train,standard_clf.predict(standard_scalered_x_train))
    standard_scalerd_accuracy_test=accuracy_score(y_test,standard_clf.predict(standard_scalered_x_test))

    standard_accuracies_train+=standard_scalerd_accuracy_train
    standard_accuracies_test+=standard_scalerd_accuracy_test

    #--------------- Min max skaliranje ----------------------#
    min_max_scaler = MinMaxScaler()
    min_max_scalerd_x_train=min_max_scaler.fit_transform(X_train)
    min_max_scalerd_x_test=min_max_scaler.transform(X_test)

    min_max_clf=SVC().fit(min_max_scalerd_x_train,y_train)
    min_max_accuracy_train=accuracy_score(y_train,min_max_clf.predict(min_max_scalerd_x_train))
    min_max_accuracy_test=accuracy_score(y_test,min_max_clf.predict(min_max_scalerd_x_test))

    minmax_accuracies_train+=min_max_accuracy_train
    minmax_accuracies_test+=min_max_accuracy_test

print(f'Prosječna preciznost train regulanih podataka : {regular_accuracies_train/iter}')
print(f'Prosječna preciznost train standardno skaliranih podataka : {standard_accuracies_train/iter}')
print(f'Prosječna preciznost train min max skaliranih podataka : {minmax_accuracies_train/iter}')
print()
print(f'Prosječna preciznost test regulanih podataka : {regular_accuracies_test/iter}')
print(f'Prosječna preciznost test standardno skaliranih podataka : {standard_accuracies_test/iter}')
print(f'Prosječna preciznost test min max skaliranih podataka : {minmax_accuracies_test/iter}')
```
